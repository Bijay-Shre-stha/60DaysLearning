```python
# A random variable X is uniformly distributed between 0 and 10. Generate 1000 random samples from this distribution and plot the histogram of these samples. Overlay the probability density function (PDF) of the uniform distribution on the histogram.

from scipy.stats import uniform


a = 0
b = 10

samples = np.random.uniform(a,b,1000)
plt.hist(samples, bins=20, density=True, alpha=0.6, color='b', edgecolor='black')

x = np.linspace(a,b,100)
pdf = uniform.pdf(x, a, b-a)
plt.plot(x, pdf, 'r', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniform Distribution')
plt.grid(True)
plt.show()
```
