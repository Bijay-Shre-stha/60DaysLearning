```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_walk(steps):
    x_coords = [0]
    y_coords = [0]
    # Directions: 1 = right, 2 = left, 3 = up, 4 = down
    for _ in range(steps):
        direction = np.random.randint(1, 5)
        if direction == 1:
            x_coords.append(x_coords[-1] + 1)
            y_coords.append(y_coords[-1])
        elif direction == 2:
            x_coords.append(x_coords[-1] - 1)
            y_coords.append(y_coords[-1])
        elif direction == 3:
            x_coords.append(x_coords[-1])
            y_coords.append(y_coords[-1] + 1)
        else:  # direction == 4
            x_coords.append(x_coords[-1])
            y_coords.append(y_coords[-1] - 1)
    return x_coords, y_coords

def plot_random_walk(x_coords, y_coords):
    plt.figure(figsize=(8, 8))
    plt.plot(x_coords, y_coords, linestyle='-', marker='o', color='b', markersize=5)
    plt.scatter(x_coords[0], y_coords[0], color='g', label='Start', zorder=5)
    plt.scatter(x_coords[-1], y_coords[-1], color='r', label='End', zorder=5)
    plt.title('Random Walk')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.legend()
    plt.show()

def main():
    steps = 50
    x_coords, y_coords = generate_random_walk(steps)
    for i in range(steps + 1):
        print(f"Step {i}: ({x_coords[i]}, {y_coords[i]})")
    plot_random_walk(x_coords, y_coords)

if __name__ == "__main__":
    main()
```
