```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt

# Logistic distribution
x = random.logistic(loc=1, scale=2, size=(2, 3))
print(x)
```

```python
# Visualization
import seaborn as sns
sns.distplot(random.logistic(size=1000), hist=False)
plt.show()
```

```python
# Difference Between Logistic and Normal Distribution

# The main difference is that logistic distribution is more peaked than normal distribution.
# The logistic distribution has a higher kurtosis value than the normal distribution.
# The logistic distribution has a wider range of values compared to the normal distribution.
# The logistic distribution is used in deep learning and neural networks.

sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')

plt.show()
```

```python
# varying position arguments

import matplotlib.pyplot as plt  
import numpy as np  
from scipy.stats import logistic

x = np.linspace(0, 5, 100)  

# Varying positional arguments  
y1 = logistic.pdf(x, 1, 3)  
y2 = logistic.pdf(x, 1, 4)  
plt.plot(x, y1, "*", x, y2, "r--")  
plt.show()
```
