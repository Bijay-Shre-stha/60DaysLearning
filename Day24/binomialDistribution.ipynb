```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# Number of trials
n = 10
p = 0.6

# Probability of getting 5 heads
prob = binom.pmf(5, n, p)
print(prob)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.show()
```

```python
# Difference between Binomial and Normal Distribution

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=False, kde=True, color='r', label='binomial')
sns.distplot(random.normal(loc=5, scale=1, size=1000), hist=False, kde=True, color='b', label='normal')

plt.show()
```

```python
# A company claims that 60% of its employees have received professional training. If you randomly select 10 employees, what is the probability that exactly 6 of them have received professional training?

n = 10
p = 0.6
k = 6
probability = binom.pmf(k, n, p)

print(f"The probability that exactly 6 out of 10 employees have received professional training is {probability:.4f}")

x = np.arange(0, n + 1)
# Calculate binomial probabilities for all outcomes
binom_probs = binom.pmf(x, n, p)

# Plotting the binomial distribution
plt.figure(figsize=(10, 6))
plt.bar(x, binom_probs, color='skyblue', edgecolor='black')
plt.plot(x, binom_probs, 'bo', label='Probability')
plt.title('Binomial Distribution (n=10, p=0.60)')
plt.xlabel('Number of employees who received training')
plt.ylabel('Probability')
plt.xticks(np.arange(0, n + 1, 1))
plt.grid(axis='y')
```
