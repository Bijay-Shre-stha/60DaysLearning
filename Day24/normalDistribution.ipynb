```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x,mean,sd):
    probability = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return probability

mean = 0
sd = 1
x = 1
print(normal_distribution(x,mean,sd))
```

```python
# Example question 

# Suppose there are 100 students in the class and in one of the mathematics tests the average marks scored by the students in the subject is 78 and the standard deviation is 25. The marks of the student follow Normal probability distribution. We can use this information to answer some questions about the student’s marks. 
# Python Code for Percentage of Students who got less than 60 marks
# Here we will use the norm() function from scipy.stats module to make the probability distribution for the population’s mean equal to 78 and the standard deviation equal to 25.

from scipy.stats import norm

# Given data
mean = 78
sd = 25
total_students = 100
score = 60

# Calculating the z-score
z_score = (score - mean) / sd

# Calculating the percentage of students who scored less than 60
students = norm.cdf(z_score)
students_percentage = students * 100
print("Percentage of students who got less than 60 marks: ", round(students_percentage, 2), "%")

# Percentage of Students who got more than 70 marks

score = 70
z_score = (score - mean) / sd
students = 1 - norm.cdf(z_score)
students_percentage = students * 100
print("Percentage of students who got more than 70 marks: ", round(students_percentage, 2), "%")

#  Percentage of Students who have scored More than 75 and less than 85

min_score = 75
max_score = 85

z_min_score = (min_score - mean) / sd
z_max_score = (max_score - mean) / sd

students_min_score = norm.cdf(z_min_score)
students_max_score = norm.cdf(z_max_score)

students_percentage = (students_max_score - students_min_score) * 100
print("Percentage of students who got more than 75 and less than 85 marks: ", round(students_percentage, 2), "%")

# Plotting the Normal Distribution

x = np.linspace(0, 150, 100)
y = norm.pdf(x, mean, sd)
plt.plot(x, y)
plt.grid(True)
plt.xlabel("Marks")
plt.ylabel("Probability Density")
plt.title("Normal Distribution of Student's Marks")
plt.show()

```
