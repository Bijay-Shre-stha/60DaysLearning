```python
# Given a dataset of wind speeds measured in meters per second (m/s) at a particular location, fit the data to a Rayleigh distribution. Then, using the fitted parameters, answer the following:

# What is the most likely wind speed (the mode) at this location?
# What is the probability that the wind speed exceeds 10 m/s?
# You can use the following dataset for wind speeds:

# wind_speeds = [2.3, 3.4, 4.5, 1.2, 5.6, 7.8, 3.4, 5.5, 6.7, 2.3, 4.5, 3.4, 2.1, 8.9, 6.5, 4.3, 3.2, 7.1, 5.4, 6.6]

# solution
from scipy.stats import rayleigh
import numpy as np
import matplotlib.pyplot as plt

wind_speeds = [2.3, 3.4, 4.5, 1.2, 5.6, 7.8, 3.4, 5.5, 6.7, 2.3, 4.5, 3.4, 2.1, 8.9, 6.5, 4.3, 3.2, 7.1, 5.4, 6.6]

# Fit the data to a Rayleigh distribution
params = rayleigh.fit(wind_speeds)
fitted_loc = params[0]
fitted_scale = params[1]

print(f"Fitted location: {fitted_loc}")
print(f"Fitted scale: {fitted_scale}")

# Calculate the mode
mode = fitted_loc + fitted_scale
print(f"The most likely wind speed (mode): {mode:.2f} m/s")

# Calculate the probability that the wind speed exceeds 10 m/s
prob_exceed_10 = 1 - rayleigh.cdf(10, loc=fitted_loc, scale=fitted_scale)
print(f"The probability that the wind speed exceeds 10 m/s: {prob_exceed_10:.4f}")

# Visualize the fitted distribution
x = np.linspace(0, 12, 1000)
pdf_fitted = rayleigh.pdf(x, loc=fitted_loc, scale=fitted_scale)

plt.figure(figsize=(10, 6))
plt.hist(wind_speeds, bins=10, density=True, alpha=0.6, color='g', label='Observed data')
plt.plot(x, pdf_fitted, 'r-', lw=2, label='Fitted Rayleigh distribution')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Probability Density')
plt.title('Fitted Rayleigh Distribution for Wind Speeds')
plt.legend()
plt.grid()
plt.show()

```
