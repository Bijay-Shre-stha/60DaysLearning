```python
data_frame = pd.read_csv('data.csv')
data_frame["Calories"].fillna(130, inplace = True)
print(data_frame)
```

```python
# Replace Using Mean
data_frame = pd.read_csv('data.csv')
x = data_frame["Calories"].mean()
data_frame["Calories"].fillna(x, inplace = True)
print(data_frame)
```

```python
# Replace Using Median

data_frame = pd.read_csv('data.csv')
x = data_frame["Calories"].median()
data_frame["Calories"].fillna(x, inplace = True)
print(data_frame)
```

```python
# Replace Using Mode

data_frame = pd.read_csv('data.csv')
x = data_frame["Calories"].mode()[0]
data_frame["Calories"].fillna(x, inplace = True)
print(data_frame)
```
