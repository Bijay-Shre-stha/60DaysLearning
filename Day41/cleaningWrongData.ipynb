```python
import pandas as pd
data_frame = pd.read_csv('data.csv')
print(data_frame)
```

```python
for x in data_frame.index:
    if data_frame.loc[x, "Duration"]>120:
        data_frame.loc[x, "Duration"] = 120

print(data_frame)
```

```python
# remove rows with missing data
for x in data_frame.index:
    if data_frame.loc[x, "Duration"]>120:
        data_frame.drop(x, inplace=True)
print(data_frame)
```
