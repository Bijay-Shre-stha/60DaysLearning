```python
# bar char for categorical features

def bar_chart(feature):
    survived = train[train['Survived'] == 1][feature].value_counts()
    dead = train[train['Survived'] == 0][feature].value_counts()
    df = pd.DataFrame([survived, dead])
    df.index = ['Survived', 'Dead']
    df.plot(kind='bar', stacked=True, figsize=(10, 5))
```

# Model Selection

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (RandomForestClassifier, GradientBoostingClassifier)
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
train.info()
```

# Cross Validation (K-fold)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors = 13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
clf = [
    ('KNeighbors', KNeighborsClassifier(n_neighbors=13)),
    ('DecisionTree', DecisionTreeClassifier()),
    ('RandomForest', RandomForestClassifier(n_estimators=13)),
    ('GaussianNB', GaussianNB()),
    ('SVC', SVC()),
    ('GradientBoosting', GradientBoostingClassifier(n_estimators=10, learning_rate=1, max_features=3, max_depth=3, random_state=10)),
    ('LogisticRegression', LogisticRegression(max_iter=200))
]
# Define a function to fit models and print scores
def model_fit():
    scoring = 'accuracy'
    for name, model in clf:
        score = cross_val_score(model, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
        print(f"Score of {name}: {round(np.mean(score) * 100, 2)}%")

# Call the function
model_fit()
```

```python
# plot model scoring

import seaborn as sns

# plot model scores
model_names_score = {
    'KNeighbors': 82.6,
    'DecisionTree': 79.58,
    'RandomForest': 80.7,
    'GaussianNB': 78.78,
    'SVC': 83.5,
    'GradientBoosting': 81.25,
    'LogisticRegression': 81.26
}

sorted_model_names_score = dict(sorted(model_names_score.items(), key=lambda item: item[1], reverse=True))

sns.set_theme(style='whitegrid')
plt.figure(figsize=(13, 7))
bar_plot = sns.barplot(x=list(sorted_model_names_score.keys()), y=list(sorted_model_names_score.values()), palette='viridis')

for index, value in enumerate(sorted_model_names_score.values()):
    bar_plot.text(index, value + 0.2, f"{value:.2f}%", ha='center', va='bottom', color='black', fontsize=12)

plt.title('Model Scores for Titanic - Machine Learning from Disaster', fontsize=16, fontweight='bold')
plt.ylabel('Accuracy Score (%)', fontsize=14)
plt.xlabel('Model Name', fontsize=14)
plt.ylim(75, 85)  

plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

### Conclusion

Based on the evaluation, the Support Vector Classifier (SVC) emerged as the top-performing model with an accuracy of 83.5%. However, it's important to consider other factors such as computational efficiency, interpretability, and potential overfitting when choosing the final model for deployment.
